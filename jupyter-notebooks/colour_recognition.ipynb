# Rozpoznawanie rzeczywistego wymiaru (liczby kanałów) obrazu

#### Importowanie bibliotek

```python
import matplotlib.pyplot as plt
import cv2
```

#### Funkcje

```python
def show_image(image, title=""):
    if image.ndim == 2:
        plt.imshow(image,cmap='gray')
    else:
        plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
```

```python
def recognition_1(image):
    return image.ndim
```

```python
def recognition_2(image):
    for row in range(image.shape[0]):
        for column in range(image.shape[1]):
            if not (image[row][column][0] == image[row][column][1] == image[row][column][2]):
                return True
    return False
```

Podsumowując, ostateczną funkcją do sprawdzania rzeczywistej liczby kanałów będzie funkcja recognition_2, która przyjmie nazwę is_color.

```python
def is_color(image):
    for row in range(image.shape[0]):
        for column in range(image.shape[1]):
            if not (image[row][column][0] == image[row][column][1] == image[row][column][2]):
                return True
    return False
```
